```python
import plotly.io as pio
pio.renderers.default = "png"
import pandas as pd
df = pd.read_csv('/content/supply_chain_data.csv')
df.head()
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
df.drop_duplicates(inplace=True)
df.fillna(method='ffill', inplace=True)
```

```python
df.describe()
```

```python
import plotly.express as px

fig = px.bar(df,
             x='Product type',
             y='Revenue generated',
             color='Product type',
             title='Total Revenue by Product Type',
             color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_layout(title_x=0.5)
fig.show()
```

```python
import plotly.express as px
fig_ship = px.scatter(df, x='Shipping times', y='Shipping costs',
                      color='Shipping carriers', size='Number of products sold',
                      title='Shipping Efficiency: Time vs Cost',
                      color_discrete_sequence=px.colors.qualitative.Pastel)
fig_ship.show()

fig_defect = px.bar(df, x='Supplier name', y='Defect rates',
                    color='Product type',
                    title='Defect Rates by Supplier and Product Type',
                    barmode='group',
                    color_discrete_sequence=px.colors.qualitative.Safe)
fig_defect.show()

fig_stock = px.pie(df, values='Stock levels', names='Product type',
                   title='Current Stock Levels Distribution',
                   hole=0.4)
fig_stock.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

df['Net Profit'] = df['Revenue generated'] - (df['Manufacturing costs'] + df['Shipping costs'])
fig1 = px.bar(df, x='Product type', y='Net Profit',
             title='Net Profitability by Product Category',
             color='Product type',
             color_discrete_sequence=['#457B9D', '#A8DADC', '#E63946'])
fig1.update_layout(title_x=0.5, template='plotly_white')
fig1.show()

fig2 = px.scatter(df, x='Manufacturing costs', y='Defect rates',
                 size='Number of products sold', color='Supplier name',
                 title='Supplier Performance: Costs vs Quality',
                 color_discrete_sequence=px.colors.qualitative.T10)
fig2.update_layout(title_x=0.5)
fig2.show()

fig3 = px.box(df, x='Location', y='Lead time',
             color='Location',
             title='Supply Chain Delay by Location',
             color_discrete_sequence=px.colors.qualitative.Pastel)
fig3.update_layout(title_x=0.5)
fig3.show()
```
